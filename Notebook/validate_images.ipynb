```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Paths to the datasets
original_images_dir = "../artifacts/rendered_images"  # Folder with original generated images
augmented_images_dir = "../artifacts/augumented_images"  # Folder with augmented images
```

```python
def validate_images(directory, num_samples=5):
    """
    Validates a dataset by randomly sampling images and displaying them.

    Args:
    - directory (str): Path to the image directory.
    - num_samples (int): Number of random images to display.

    Returns:
    - None
    """
    # List all image files in the directory
    image_files = [f for f in os.listdir(directory) if f.endswith(".png") or f.endswith(".jpg")]
    
    # Check if the directory contains images
    if not image_files:
        print(f"No images found in {directory}. Please check the path or generate the images.")
        return
    
    # Randomly sample images
    sampled_files = random.sample(image_files, min(num_samples, len(image_files)))
    print(f"Displaying {len(sampled_files)} images from {directory}...")

    # Plot the images
    plt.figure(figsize=(15, 10))  # Set the figure size
    for i, image_file in enumerate(sampled_files):
        # Load the image
        image_path = os.path.join(directory, image_file)
        image = Image.open(image_path)

        # Display the image
        plt.subplot(1, len(sampled_files), i + 1)  # Create a subplot
        plt.imshow(image)
        plt.axis("off")  # Turn off axes
        plt.title(image_file)  # Display the image name

    plt.show()  # Show the images

# Validate the original generated images
print("Validating original generated images:")
validate_images(original_images_dir, num_samples=5)

# Validate the augmented images
print("\nValidating augmented images:")
validate_images(augmented_images_dir, num_samples=5)
```
